## Ranking the States by avg annual snowfall

```python
df.groupby(["State"])['Avg annual snowfall (in)'].mean().sort_values(ascending=False)
```

```python
ordinal_labels=df.groupby(["State"])['Avg annual snowfall (in)'].mean().sort_values(ascending=False).index
ordinal_labels
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df['State_mean_avg_snowfall_rank']=df['State'].map(ordinal_labels2)
df.head()
```
